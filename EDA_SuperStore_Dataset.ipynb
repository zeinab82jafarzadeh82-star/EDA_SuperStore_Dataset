برای اینکه مشتری دائم را به طور تخصصی و پیشرفته تر شناسایی کنیم می توانیم از روش

RFM

استفاده کنیم

```python
#RFM
#R -> RECENCY  = آخرین خرید مشتری
#F -> FREQUENCY = تعداد دفعات خرید
#M -> MONETARY = مجموع مبلغ خرید

# تبدیل کنیم(datetime)در ابتدا باید تاریخ سفارش را به
# است(object)زیرا تاریخ سفارش از نوع
superStore_df['Order Date']=pd.to_datetime(superStore_df['Order Date'])

#تعیین آخرین تاریخ برای تاریخ مرجع
reference_date=superStore_df['Order Date'].max()   #'2022-12-30 00:00:00'

#making rfm table
rfm=superStore_df.groupby('Customer ID').agg({
    'Order Date': lambda x: (reference_date - x.max()).days, #Recency
    'Order ID' : 'nunique',                                  #Frequency
    'Sales' : 'sum'                                          #Monetary
})

rfm.columns = ['Recency','Frequency','Monetary']
rfm=rfm.reset_index()

#نمره دهی به هر شاخص
#qcut() function cuts and classificates om a way that we tell
rfm['R_score']=pd.qcut(rfm['Recency'], 5 , labels=[5,4,3,2,1] ) # Recency کمتر → نمره بیشتر
rfm['F_score']=pd.qcut(rfm['Frequency'].rank(method='first'), 5 , labels=[1,2,3,4,5])
rfm['M_score']=pd.qcut(rfm['Monetary'], 5 , labels=[1,2,3,4,5])

rfm

#making RFM Segment
rfm['RFM_segment']=rfm['R_score'].astype(str)+rfm['F_score'].astype(str)+rfm['M_score'].astype(str)
rfm['RFM_score']=rfm[['R_score','F_score','M_score']].astype(int).sum(axis=1)

#classification customers by RFM_score
rfm['Customer_Level']=rfm['RFM_score'].apply(lambda x :
       'Best Customer' if x>=13 else
       'Loyal Customer' if x>=10 else
       'Potential Customer' if x>=7 else
       'Needs Attention')

#draw barplot for Customer Level Distribution
plt.title('Customer Level Distribution')
rfm['Customer_Level'].value_counts().plot(kind='bar')
plt.xlabel('Customer Level')
plt.show()

rfm
```
